```python
import torch

def jaccard_similarity(tensor1, tensor2):
    # 将二维张量展平为一维
    set1 = set(tensor1.view(-1).tolist())
    set2 = set(tensor2.view(-1).tolist())
    
    # 计算交集和并集
    intersection = len(set1 & set2)
    union = len(set1 | set2)
    
    # 计算Jaccard相似度
    jaccard_similarity = intersection / union

    return jaccard_similarity

# 示例
tensor1 = torch.tensor([[1, 2], [3, 4]])
tensor2 = torch.tensor([[2, 3], [4, 5]])

similarity = jaccard_similarity(tensor1, tensor2)
print(similarity)
```
